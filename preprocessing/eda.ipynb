# EDA Notebook

Some simple EDA done on the dataset; mainly just to find out which are the most popular airport locations (both `Origin` and `Dest`) to be used for the historical weather scraping.

```python
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
```

```python
origin_count = Counter()
dest_count = Counter()

chunksize = 100000

for chunk in pd.read_csv("../datafiles/airline.csv.shuffle", 
                         encoding="latin-1",
                         usecols=["Origin", "Dest"],
                         chunksize=chunksize,
                         low_memory=True):

    origin_count.update(chunk["Origin"].dropna())
    dest_count.update(chunk["Dest"].dropna())
```

```python
df_origin = pd.DataFrame(origin_count.most_common(), 
                         columns=["Airport", "Count"])
```

```python
df_dest = pd.DataFrame(dest_count.most_common(), 
                       columns=["Airport", "Count"])
```

```python
df_dest.head(10)
```

```python
# Visualise the top 20 origin airports
top_origin_20 = df_origin.head(20)

plt.figure(figsize=(12, 6))
plt.bar(top_origin_20["Airport"], top_origin_20["Count"])
plt.xticks(rotation=45, ha='right')
plt.title("Top 20 Origin Airports by Departures")
plt.xlabel("Airport (IATA)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Visualise the top 10 origin airports
top_origin_10 = df_origin.head(10)

plt.figure(figsize=(12, 6))
plt.bar(top_origin_10["Airport"], top_origin_10["Count"])
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Origin Airports by Departures")
plt.xlabel("Airport (IATA)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Visualise the top 20 destination airports
top_dest_20 = df_dest.head(20)

plt.figure(figsize=(12, 6))
plt.bar(top_dest_20["Airport"], top_dest_20["Count"])
plt.xticks(rotation=45, ha='right')
plt.title("Top 20 Destination Airports by Arrivals")
plt.xlabel("Airport (IATA)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
