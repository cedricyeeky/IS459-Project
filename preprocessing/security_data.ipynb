# Preprocesing for Security Data

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/globalterrorism_raw.csv', encoding='latin1')
df.head()
```

```python
df = df[['iyear', 'imonth', 'iday', 'country_txt']]
df.head()
```

```python
df['country_txt'].unique()
```

```python
# Filter for United States
df_us = df[
    (df['country_txt'] == 'United States') &
    (df['iyear'] >= 1987) &
    (df['iyear'] <= 2008)
].copy()

# Create a date column
df_us['date'] = pd.to_datetime(
    dict(year=df_us.iyear, month=df_us.imonth, day=df_us.iday),
    errors='coerce'
)

df_us.head()
```

```python
df_us.to_csv('../data/us_terrorism_1987_2008.csv', index=False)
```
